```python
import folium
from folium import IFrame
from folium.features import RegularPolygonMarker 

center =31.57, -84.4

table = """
<!DOCTYPE html>
<html>
<head>
<style>
table {{
    width:100%;height:100%
}}
table, th, td {{
    border: 1px solid black;
    border-collapse: collapse;
}}
th, td {{
    padding: 5px;
    text-align: left;
}}
table#t01 tr:nth-child(odd) {{
    background-color: #eee;
}}
table#t01 tr:nth-child(even) {{
   background-color:#fff;
}}
</style>
</head>
<body>

<table id="t01">
  <tr>
    <td>Id</td>
    <td>{}</td>
  </tr>
  <tr>
    <td>Acquistion Date</td>
    <td>{}</td>
  </tr>
  <tr>
    <td>Satellite</td>
    <td>{}</td>
  </tr>
</table>
</body>
</html>
""".format

# Define the Map
map = folium.Map(
    location=center,
    zoom_start=10,
    control_scale=True
)

def style_function(feature):
    return {
        lambda x: {'fillColor': '#ffffff' if x['properties']['constellation']=='SPOT' else '#00ff00'}
    }

def highlight_function(feature):
    return {
        'fillColor': '#ffaf00'
  #      'color': 'green',
  #      'weight': 3,
  #      'dashArray': '5, 5'
    }

# Define BBOX styling function
style_function = lambda x: {'fillColor': '#ffffff' if x['properties']['constellation']=='SPOT' else '#00ff00'}

# Define the GeoJSON Layer
gj = folium.features.GeoJson(results,style_function=style_function,highlight_function=highlight_function,overlay=False)

width, height = 410,130

fg = folium.FeatureGroup(name='OneAtlas')

for f in results['features']:
    iframe = IFrame(table(f['id'], f['properties']['acquisitionDate'], f['properties']['satellite']), width=width, height=height)
    fg.add_child(RegularPolygonMarker(location=f['properties']['bbox'],popup=(folium.Popup(iframe,max_width=2650))))
    
# Add GeoJSON to the Map
map.add_child(fg)
map.add_child(gj)

# Display Map
map
```
